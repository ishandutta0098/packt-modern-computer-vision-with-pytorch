# Operations on Tensors

```python
import torch
```

```python
# Multiplication

x = torch.tensor([
    [10, 20, 30], 
    [1, 2, 3]
    ])

y = x * 10

print(y)
```

```python
# Addition

x = torch.tensor([
    [10, 20, 30], 
    [1, 2, 3]
    ])

y = x.add(10)

print(y)
```

```python
# Reshape using view

y = torch.tensor([1,2,3,4])
print(y.shape)
print(y)
print()

y = y.view(4,1)
print(y.shape)
print(y)
```

```python
# Reshape using squeeze 
# We provide axis index that we want to remove and is applicable only when the axis we want to remove has
# only one item in that dimension

x = torch.randn(10, 1, 10)
print(x.shape)

z1 = torch.squeeze(x, 1)
print(z1.shape)

# or

z2 = x.squeeze((1))
print(z2.shape)
```

```python
# Reshape using unsqueeze

x = torch.randn(10, 10)
print(x.shape)

z1 = x.unsqueeze(0)
print(z1.shape)
```

```python
# Unsqueeze with None

x = torch.randn(10, 10)
print(x.shape)

z2, z3, z4 = x[None], x[:, None], x[:, :, None]
print(z2.shape, z3.shape, z4.shape)
```

```python
# Matrix Multiplication using @

print(x @ y)
```

```python
# Concatenate

x = torch.randn(10, 10, 10)
z = torch.cat([x, x], axis = 0)

print("Cat axis: 0")
print(x.shape)
print(z.shape)

z = torch.cat([x, x], axis = 1)

print("\nCat axis: 1")
print(x.shape)
print(z.shape)
```

```python
# Maximum value in a tensor

x = torch.arange(25).reshape(5, 5)
print(x)
print("\nMax: ", torch.max(x))
```

```python
# Max values along rows with their row index (argmax)
m, argmax = x.max(dim=0)
print(m)
print(argmax)
```

```python
# Max values along columns with their columns index (argmax)
m, argmax = x.max(dim=1)
print(m)
print(argmax)
```

```python
# Permute Tensor

x = torch.randn(10, 20, 30)
z = x.permute(2,0,1)
print(x.shape)
print(z.shape)
```
