# Inspecting Color Images

```python
# Imports

import cv2
import matplotlib.pyplot as plt
```

```python
# Load Image
img = cv2.imread('D:/book-implementations/packt-modern-computer-vision-with-pytorch/3-building-deep-nn-with-pytorch/images/Hemanvi.jpeg')

# Crop Image
img = img[50:250, 40:240, :]

# Color
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Plot image
plt.imshow(img)

print(img.shape)
```

```python
# Bottom right 3x3 array

crop = img[-3:, -3:, :]

print(crop)

plt.imshow(crop)
```
